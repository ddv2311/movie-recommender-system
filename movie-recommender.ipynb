```python
import ast
```

```python
## make a list which has tags for movies
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
##In cast only top three actors are needed tobe considered
def convert(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
##here the name from the dictionary to be extracted is one with job director 

def find_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies.head(100)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
## white spaces are needed to be removed from all the list for movie recommendation
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
## make a new column named tags which will be concatination of overview,genres,keywords,cast,crew

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
df1 = movies[['movie_id','title','tags']]
```

```python
##convert list to string
df1['tags'] = df1['tags'].apply(lambda x:" ".join(x))
```

```python
df1['tags'] = df1['tags'].apply(lambda x:x.lower())
```

```python
##here all are converted to root word with use of stemming
##here cosine distance is used to measure the distance between the vectors as euclidan distance fails for the higher dimension
from sklearn.metrics.pairwise import cosine_similarity


```

```python
similarity = cosine_similarity(vectors)
```

now the recommend will be done in the following way-:
1)user searches a movie 
2) the index of that movie is to be found
3)sort the similarity matix of that movie in descending order and fetch top  5 movies

```python
def recommend(movie):
    movie_index = df1[df1['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(df1.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
